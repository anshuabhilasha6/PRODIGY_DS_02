```python
#Importing necessary libaries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading the file
data=pd.read_csv("test.csv")
```

```python
#fetching top 5 data
data.head()
```

```python
#fetching data from bottom
data.tail()
```

```python
#using describe to get the mean,count,std,etc
data.describe()
```

```python
#using info function to know about all null values and the data types
data.info()
```

```python
#to know about the blank columns and rows
data.isnull().sum()
```

```python
#here we got to know that fare has 1 null value,cabin has 327 null value,age has 86 null values
#CLEANING OF NULL VALUE
data.dropna(subset=["Embarked"],inplace=True)
data["Cabin"].fillna("Unknown",inplace=True)
data["Age"].fillna(data["Age"].mean(),inplace=True)
```

```python
#checking the changes
data.isnull().sum()
```

```python
#checking for the duplicate values
data.duplicated().sum()
```

```python
plt.figure(figsize=(6,3))
sns.histplot(data["Age"],kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
#this is our histogram plot.
```

```python
#creating a countplot
plt.figure(figsize=(6,3))
sns.countplot(data=data,x="Sex",hue="Sex")
plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Survived",loc="upper right")
plt.show()
```

```python
#creating a scatterplot on the basis of age
plt.figure(figsize=(6,3))
sns.scatterplot(data=data,x="Age",y="Fare",hue="Age")
plt.title("Scatterplot AGE VS FARE")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend(title="Survived")
plt.show()
```
